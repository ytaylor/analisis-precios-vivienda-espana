```python
import pandas as pd

# Cargar el csv
df = pd.read_csv("files/Valor_Tasado.csv")
```

```python
df.columns
```

```python
df_compraventa = pd.read_csv("files/Compraventa.csv", encoding="utf-8", sep=";")
```

```python
print(df_compraventa.columns)
```

```python
df_compraventa["Comunidades y Ciudades Autónomas"].unique()
```

```python
df_compraventa["Comunidades y Ciudades Autónomas"] = df_compraventa["Comunidades y Ciudades Autónomas"].str.replace(r'^\d+\s', '', regex=True)

df_compraventa["Comunidades y Ciudades Autónomas"].unique()
```

```python
# Diccionario de reemplazos: clave = nombre original, valor = nombre normalizado
replacements = {
    'Asturias (Principado de )': 'Asturias, Principado de',
    'Balears (Illes)': 'Balears, Illes',
    'Comunidad Valenciana': 'Comunitat Valenciana',
    'Valencia/València': 'Comunitat Valenciana',
    'Madrid (Comunidad de)': 'Madrid, Comunidad de',
    'Murcia (Región de)': 'Murcia, Región de',
    'Navarra (Com. Foral de)': 'Navarra, Comunidad Foral de',
    'Navarra (Comunidad Foral de)': 'Navarra, Comunidad Foral de',
    'Rioja (La)': 'Rioja, La',
    # Añade más reemplazos si es necesario
}
```

```python
df['Comunidad_Autonoma'] = df['Comunidad_Autonoma'].replace(replacements)
df["Comunidad_Autonoma"].unique()
```

```python
ccaa = [
   'Andalucía', 
   'Aragón', 
   'Asturias, Principado de',
    'Balears, Illes', 
    'Canarias', 
    'Cantabria', 
    'Castilla y León',
    'Castilla-La Mancha',
    'Cataluña', 
    'Comunitat Valenciana',
    'Extremadura', 
    'Galicia', 
    'Madrid, Comunidad de',
    'Murcia, Región de', 
    'Navarra, Comunidad Foral de', 
    'País Vasco',
    'Rioja, La', 
    'Ceuta', 
    'Melilla']
```

```python
df_ccaa = df[df["Comunidad_Autonoma"].isin(ccaa)].copy()
df_ccaa.head(10)
```

```python
df_ccaa["Comunidad_Autonoma"].unique()
```

```python
df_provincias = df[~df["Comunidad_Autonoma"].isin(ccaa)].copy()
df_provincias.head(10)
```

```python
## Guardar los nuevos csv
df_ccaa.to_csv("valor_m2_ccaa.csv", index=False)
df_provincias.to_csv("valor_m2_provincias.csv", index=False)
```
