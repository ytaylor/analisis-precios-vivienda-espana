# Análisis de Precios de Vivienda en España

Este notebook utiliza pandas y matplotlib para leer y analizar datos de precios de vivienda en España.

## Objetivos:
- Leer ficheros CSV y Excel con pandas
- Explorar y analizar datos de precios de vivienda
- Visualizar datos con matplotlib
- Responder preguntas sobre el mercado inmobiliario

## 1. Importar librerías necesarias

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Configuración para mejorar la visualización
plt.style.use('seaborn-v0_8-darkgrid')
%matplotlib inline
```

## 2. Leer datos desde archivo CSV

```python
# Leer el archivo CSV con datos de vivienda
df_vivienda = pd.read_csv('datos_vivienda.csv')

# Mostrar las primeras filas
print("Primeras 5 filas del dataset:")
df_vivienda.head()
```

## 3. Exploración inicial de los datos

```python
# Información general del dataset
print("Información del dataset:")
print(df_vivienda.info())
print("\n" + "="*50 + "\n")

# Estadísticas descriptivas
print("Estadísticas descriptivas:")
df_vivienda.describe()
```

## 4. Preguntas y Análisis sobre los Datos

### Pregunta 1: ¿Cuál es el precio medio por m² en cada ciudad?

```python
# Calcular precio medio por m² por ciudad
precio_medio_ciudad = df_vivienda.groupby('Ciudad')['Precio_m2'].mean().sort_values(ascending=False)

print("Precio medio por m² en cada ciudad:")
print(precio_medio_ciudad)

# Visualizar con un gráfico de barras
plt.figure(figsize=(12, 6))
precio_medio_ciudad.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Precio Medio por m² por Ciudad', fontsize=16, fontweight='bold')
plt.xlabel('Ciudad', fontsize=12)
plt.ylabel('Precio por m² (€)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

### Pregunta 2: ¿Cuántas viviendas hay por tipo (Piso vs Casa)?

```python
# Contar viviendas por tipo
viviendas_tipo = df_vivienda['Tipo_Vivienda'].value_counts()

print("Distribución por tipo de vivienda:")
print(viviendas_tipo)

# Crear un gráfico de pastel
plt.figure(figsize=(8, 8))
plt.pie(viviendas_tipo.values, labels=viviendas_tipo.index, autopct='%1.1f%%', 
        startangle=90, colors=['#ff9999', '#66b3ff'])
plt.title('Distribución de Viviendas por Tipo', fontsize=16, fontweight='bold')
plt.axis('equal')
plt.show()
```

### Pregunta 3: ¿Existe relación entre la superficie y el número de habitaciones?

```python
# Gráfico de dispersión
plt.figure(figsize=(10, 6))
plt.scatter(df_vivienda['Superficie_m2'], df_vivienda['Num_Habitaciones'], 
            c=df_vivienda['Precio_m2'], cmap='viridis', s=100, alpha=0.6, edgecolors='black')
plt.colorbar(label='Precio por m² (€)')
plt.title('Relación entre Superficie y Número de Habitaciones', fontsize=16, fontweight='bold')
plt.xlabel('Superficie (m²)', fontsize=12)
plt.ylabel('Número de Habitaciones', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Calcular correlación
correlacion = df_vivienda['Superficie_m2'].corr(df_vivienda['Num_Habitaciones'])
print(f"\nCoeficiente de correlación: {correlacion:.3f}")
```

### Pregunta 4: ¿Cómo ha evolucionado el precio medio según el año de construcción?

```python
# Agrupar por año de construcción
precio_por_año = df_vivienda.groupby('Año_Construccion')['Precio_m2'].mean().sort_index()

# Gráfico de línea
plt.figure(figsize=(12, 6))
plt.plot(precio_por_año.index, precio_por_año.values, marker='o', 
         linewidth=2, markersize=8, color='green')
plt.title('Evolución del Precio Medio por m² según Año de Construcción', 
          fontsize=16, fontweight='bold')
plt.xlabel('Año de Construcción', fontsize=12)
plt.ylabel('Precio Medio por m² (€)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

### Pregunta 5: ¿Cuáles son las 5 ciudades más caras y más baratas?

```python
# Top 5 ciudades más caras
print("Top 5 ciudades más caras (precio medio por m²):")
print(precio_medio_ciudad.head())

print("\n" + "="*50 + "\n")

# Top 5 ciudades más baratas
print("Top 5 ciudades más baratas (precio medio por m²):")
print(precio_medio_ciudad.tail())
```

## 6. Análisis Adicional: Comparación por Provincia

```python
# Análisis por provincia
provincia_stats = df_vivienda.groupby('Provincia').agg({
    'Precio_m2': ['mean', 'min', 'max'],
    'Superficie_m2': 'mean',
    'Ciudad': 'count'
}).round(2)

provincia_stats.columns = ['Precio_Medio', 'Precio_Min', 'Precio_Max', 'Superficie_Media', 'Num_Viviendas']
provincia_stats = provincia_stats.sort_values('Precio_Medio', ascending=False)

print("Estadísticas por Provincia:")
provincia_stats
```

## 7. Resumen y Conclusiones

```python
print("RESUMEN DEL ANÁLISIS")
print("=" * 60)
print(f"Total de viviendas analizadas: {len(df_vivienda)}")
print(f"Precio medio general por m²: {df_vivienda['Precio_m2'].mean():.2f} €")
print(f"Precio máximo por m²: {df_vivienda['Precio_m2'].max():.2f} €")
print(f"Precio mínimo por m²: {df_vivienda['Precio_m2'].min():.2f} €")
print(f"Superficie media: {df_vivienda['Superficie_m2'].mean():.2f} m²")
print(f"Número medio de habitaciones: {df_vivienda['Num_Habitaciones'].mean():.2f}")
print("\nCiudad más cara: {}".format(precio_medio_ciudad.index[0]))
print("Ciudad más barata: {}".format(precio_medio_ciudad.index[-1]))
```
